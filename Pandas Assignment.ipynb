```python
import numpy as np
import pandas as pd
import glob
import matplotlib.pyplot as plt
```

```python
us_census=glob.glob('states*.csv')
```

```python
lst=[]
for i in us_census:
    a=pd.read_csv(i)
    lst.append(a)
```

```python
us_census=pd.concat(lst,ignore_index=True)
```

```python
us_census
```

```python
us_census.columns
```

```python
us_census.dtypes
```

```python
us_census.drop(['Unnamed: 0'],axis=1,inplace=True)
us_census.head()
```

```python
us_census['Income']=us_census.Income.replace('\$','',regex=True)
```

```python
us_census['Income']=us_census.Income.astype('int64')
us_census.dtypes
```

```python
us_census['GenderPop']=us_census.GenderPop.replace('M','',regex=True)
us_census['GenderPop']=us_census.GenderPop.replace('F','',regex=True)
us_census.head()
```

```python
us_census[['MalePop','FemalePop']]=us_census.GenderPop.str.split('_',expand=True)
```

```python
us_census['MalePop']=us_census.MalePop.astype('int')

us_census.head()
```

```python
us_census.FemalePop=(us_census['TotalPop']-us_census['MalePop'])
us_census
```

```python
plt.xlabel('Female')
plt.ylabel('Income')
plt.scatter(us_census.FemalePop,us_census.Income)

plt.show()
```

```python
us_census.duplicated()
us_census.drop_duplicates()
```

```python
plt.xlabel('Female')
plt.ylabel('Income')
plt.scatter(us_census.FemalePop,us_census.Income)

plt.show()
```

```python
us_census.dtypes
```

```python
us_census.replace('\%','',regex=True,inplace=True)
us_census.head()
```

```python
us_census[['Hispanic','White','Black','Native','Asian','Pacific']]=us_census[['Hispanic','White','Black','Native','Asian','Pacific']].astype('float64')
us_census.dtypes
```

```python
us_census.Pacific.isnull().sum()
```

```python
us_census['Pacific']=100-(us_census.Hispanic+us_census.White+us_census.Black+us_census.Native+us_census.Asian)
```

```python
hispanicPop = us_census['Hispanic']
whitePop = us_census['White']
blackPop = us_census['Black']
nativePop = us_census['Native']
asianPop = us_census['Asian']
pacificPop = us_census['Pacific']

mybins = [0, 10, 15, 20, 25, 30, 35, 40, 45, 50, 55]

plt.hist([hispanicPop, whitePop, blackPop, nativePop, asianPop, pacificPop],
         color=['blue', 'yellow', 'red', 'black', 'green', 'orange'],
         label=['Hispanic', 'White', 'Black', 'Native', 'Asian', 'Pacific'],
         bins=mybins, edgecolor='black')

plt.title('Distribution of Races in US States')
plt.xlabel('Races')
plt.ylabel('Count')
plt.show()
```
